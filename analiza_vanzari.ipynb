```python
import pandas as pd

data = {
    'luna': ['Ian','Ian','Ian','Feb','Feb','Feb','Mar','Mar','Mar'],
    'produs': ['Laptop','Telefon','Tabletă','Laptop','Telefon','Tabletă','Laptop','Telefon','Tabletă'],
    'vanzari': [15000, 8000, 5000, 12000, 9500, 4500, 17000, 7000, 6000],
    'cantitate': [3, 10, 8, 2, 12, 7, 4, 9, 10]
}

df = pd.DataFrame(data)
df.to_csv('vanzari.csv', index=False)
print("CSV salvat!")
print(df)
```

```python
# Citim CSV-ul salvat
df = pd.read_csv('vanzari.csv')

# Total vânzări per produs
print("=== Total vânzări per produs ===")
print(df.groupby('produs')['vanzari'].sum())

# Cel mai bun produs
cel_mai_bun = df.groupby('produs')['vanzari'].sum().idxmax()
print(f"\n🏆 Cel mai bun produs: {cel_mai_bun}")

# Media vânzărilor per lună
print("\n=== Media vânzărilor per lună ===")
print(df.groupby('luna')['vanzari'].mean())

# Luna cu cele mai mari vânzări
cea_mai_buna_luna = df.groupby('luna')['vanzari'].sum().idxmax()
print(f"\n📅 Cea mai bună lună: {cea_mai_buna_luna}")
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Grafic 1 - Total vânzări per produs
total_produs = df.groupby('produs')['vanzari'].sum()
total_produs.plot(kind='bar', ax=axes[0],
                  color=['steelblue', 'coral', 'mediumseagreen'])
axes[0].set_title('Total vânzări per produs')
axes[0].set_xlabel('Produs')
axes[0].set_ylabel('Vânzări (RON)')
axes[0].tick_params(axis='x', rotation=0)

# Grafic 2 - Evoluție vânzări per lună
total_luna = df.groupby('luna')['vanzari'].sum()
total_luna.plot(kind='line', ax=axes[1],
                color='mediumpurple', marker='o', linewidth=2)
axes[1].set_title('Evoluție vânzări pe luni')
axes[1].set_xlabel('Luna')
axes[1].set_ylabel('Vânzări (RON)')

plt.tight_layout()
plt.savefig('raport_vanzari.png', dpi=150)
plt.show()
print("✅ Grafic salvat!")
```
